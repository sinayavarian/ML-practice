# Summarization
## This notebook outlines the concepts behind Text Summarization

 concept of capturing very important gist of a long piece of text

### Types of Summarization
- 1. **Extractive Summarization**
    - Select sentences from the corpus that best represent the text
    - Arrange them to form a summary
- 2. **Abstractive Summarization**
    - Captures the very important sentences from the text
    - Paraphrases them to form a summary

## TextRank (Graph-Based)
- Builds a **graph** where each sentence is a node.  
- Connects sentences with edges weighted by how similar they are.  
- Then runs a “vote” process like Google’s PageRank:
  - A sentence is important if it’s similar to other important sentences.
- The algorithm iteratively spreads importance across the network until scores stabilize.

Think of each sentence as a web page. If many sentences “link” to it (i.e., are similar), it must be central and important.

**Good for:** general summaries of well-structured text, balanced coverage.

## LexRank (Also Graph-Based, but Uses Thresholding)
- Very similar to TextRank. Also a graph-based algorithm.  
- But it uses **TF-IDF** vectors and a **similarity threshold** to decide whether to connect two sentences (e.g., only connect if similarity > 0.1).  
- Then it runs PageRank on that graph.  
- TextRank uses continuous edge weights; LexRank often uses binary connections (connected or not).  
- LexRank focuses more on **sentence centrality** based on similarity clusters.

**Good for:** long, information-dense documents (like news or Wikipedia).

### LexRankSummarizer

```python
lex_summary_text = lexSummarizer(doc, 5)
lex_summary_text
```

## Luhn (Oldest, Keyword-Frequency Method)
- Counts how often each word appears (ignores stopwords).  
- Sentences with many **high-frequency words** close together are considered important.  
- Gives higher scores to sentences dense in significant words.  

Think of highlighting repeated important words. The sentences containing them most densely are probably summary-worthy.

**Good for:** simple, factual text (like reports or definitions).  
**Weak for:** creative or varied language (since it relies on repetition).

### LuhnSummarizer

```python
luhn_summary_text = luhnSummarizer(doc, 5)
luhn_summary_text
```

### LsaSummarizer

## LSA (Latent Semantic Analysis)
- Turns the document into a **matrix** of words × sentences (how many times each word appears in each sentence).  
- Uses **Singular Value Decomposition (SVD)** (a math trick from linear algebra) to find hidden “topics” in that matrix.  
- Ranks sentences by how much they represent the main latent topics.  

**Good for:** structured text with clear topics or scientific writing.

```python
lsa_summary_text = lsaSummarizer(doc, 5)
lsa_summary_text
```

### Scrape the text

```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
def get_page(url):
    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/122.0 Safari/537.36"
        )
    }
    res = requests.get(url, headers=headers, timeout=10)
    res.raise_for_status()  # raise an error if 404/403/etc
    soup = BeautifulSoup(res.text, 'html.parser')
    return soup
```

```python
def collect_text(soup):
    # Collect all paragraph text, skipping navigation / references
    paragraphs = soup.find_all('p')
    text = "\n".join(p.get_text() for p in paragraphs if p.get_text().strip())
    return text
```

```python
url = " https://medium.com/@subashgandyer/papa-what-is-a-neural-network-c5e5cc427c7"
```

```python
text = collect_text(get_page(url))
text
```

## Gensim's `summarize()`

1. **Preprocessing**
   - The text is split into sentences.  
   - Stopwords are removed and words are stemmed.  
   - Each sentence is represented as a vector of its important words.

2. **Graph Construction**
   - Every sentence becomes a **node** in a graph.  
   - **Edges** connect sentences based on how similar they are (using word overlap or cosine similarity).

3. **Ranking (TextRank)**
   - The algorithm applies a ranking process similar to **PageRank**:
     - A sentence is considered important if it is similar to many other important sentences.
   - The importance scores stabilize after several iterations.

4. **Sentence Selection**
   - Sentences are sorted by importance.  
   - The top ones are chosen according to either:
     - A **ratio** (fraction of original text), or  
     - A **word_count** (specific summary length).

5. **Summary Assembly**
   - The selected sentences are re-ordered in their original order to form a coherent summary.




**Good for:** long, well-structured text (articles, essays, Wikipedia pages)
